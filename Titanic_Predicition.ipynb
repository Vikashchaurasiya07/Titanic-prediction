```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Titanic-Dataset.csv')
```

```python
data.dropna()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data.isnull().sum()
```

```python
data['Embarked'] =data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns=['Cabin'])
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data['Age'].hist()
```

```python
data['Age'].describe()
```

```python
data.shape
```

```python
data.head()
```

```python
data['Sex'] = data['Sex'].map({'male':0, 'female':1})
```

```python
data = pd.get_dummies(data, columns=['Embarked'], drop_first= True)
```

```python
data.head()
```

```python
data['Title'] = data['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
```

```python
data['Title'] = data['Name'].str.extract('([A-Za-z]+)\\.', expand=False)
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns=['Title'], drop_first=True)
```

```python
data.head()
```

```python
x = data.drop(columns=['PassengerId','Name','Ticket','Survived'])
y = data['Survived']
```

```python
data.head()
```

```python
X = data.drop(columns=['PassengerId','Name','Ticket','Survived'])
Y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000, solver='saga', verbose=1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Alternative visualizations: confusion matrix heatmap + counts comparison
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
# Ensure arrays
y_true = y_test.values if hasattr(y_test, 'values') else np.array(y_test)
y_pr = np.array(y_pred)
# Confusion matrix (heatmap using matplotlib)
cm = confusion_matrix(y_true, y_pr)
plt.figure(figsize=(5,4))
plt.imshow(cm, cmap='Blues')
plt.colorbar()
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha='center', va='center', color='white' if cm[i, j] > cm.max()/2 else 'black')
plt.xticks([0,1], ['Died','Survived'])
plt.yticks([0,1], ['Died','Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()
# Comparison bar chart of counts
df_comp = pd.DataFrame({'Actual': y_true, 'Predicted': y_pr})
counts_actual = df_comp['Actual'].value_counts().sort_index()
counts_pred = df_comp['Predicted'].value_counts().sort_index()
ind = np.arange(len(counts_actual))
width = 0.35
plt.figure(figsize=(6,4))
plt.bar(ind - width/2, counts_actual, width, label='Actual', color='C0')
plt.bar(ind + width/2, counts_pred, width, label='Predicted', color='C1')
plt.xticks(ind, ['Died','Survived'])
plt.ylabel('Count')
plt.title('Actual vs Predicted counts')
plt.legend()
plt.tight_layout()
plt.show()
# Print mismatch summary
mismatch_idx = np.where(y_true != y_pr)[0]
print(f'Mismatches: {len(mismatch_idx)} samples')
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))   
```
